```python
from sklearn.cluster import KMeans

import numpy as np
import pandas as pd
```

```python
np.random.seed(42)
data1 = np.random.randn(100,2) + [2, 2]
data2 = np.random.randn(100,2) + [-2, -2]
data3 = np.random.randn(100,2) + [2, -2]

X = np.vstack([data1, data2, data3])
```

```python
import matplotlib.pyplot as plt

plt.scatter(X[:, 0], X[:, 1])
```

```python
kmeans = KMeans(n_clusters=2, random_state=0)
kmeans.fit(X)
```

```python
labels = kmeans.labels_
labels
```

```python
centroids = kmeans.cluster_centers_
centroids
```

```python
plt.scatter(X[:, 0], X[:, 1], c=labels)
# centroids
plt.scatter(centroids[:, 0], centroids[:,1], c='red', marker='x', s=200, label='centroids')
plt.show()
```

```python
new_data = np.array([[0, 0], [3, 3], [-3, -3], [2, -3]])
new_labels = kmeans.predict(new_data)
new_labels
```

```python
plt.scatter(X[:, 0], X[:, 1], c=labels)
# centroids
plt.scatter(centroids[:, 0], centroids[:,1], c='red', marker='x', s=200, label='centroids')
# 테스트 데이터
plt.scatter(new_data[:,0], new_data[:,1], s=200, c=new_labels, marker='s', cmap='cool')
plt.show()
```

### DBSCAN

```python
from sklearn.datasets import make_blobs  # 부족한 데이터 생성용 패키지 

 
#중심점 설정 > 데이터 생성을  위한 > dbscan에 설정하는 것은 아님. 
centers = [[1, 1], [-1, -1], [1, -1]] 

# 실습용 데이터 생성하기 
X, labels_true = make_blobs(n_samples=750, centers=centers, cluster_std=0.4, random_state=0) 
X.shape, labels_true.shape 
```

```python
from sklearn.preprocessing import StandardScaler 

X_scaled = StandardScaler().fit_transform(X) 
X_scaled 
```

```python
import numpy as np 
from sklearn.cluster import DBSCAN 
from sklearn import metrics 

# DBSCAN 모델 정의 및 학습하기 
db = DBSCAN(eps=0.3, min_samples=10) 
db_labels = db.fit_predict(X_scaled)
db_labels[:100]
```

```python
# 생성된 클러스터의 종류
unique_db_labels = set(db_labels)
```

```python
# 노이즈의 갯수
list(db_labels).count(-1)
```

```python
# 결과 시각화
# 색상 (스펙트럼) 설정
colors = [plt.cm.Spectral(each) for each in np.linspace(0, 1, len(unique_db_labels))]
 
for k, col in zip(unique_db_labels, colors):
    if k == -1:
        # 노이즈는 검은색으로 표시
        col = [0, 0, 0, 1]
 
    # 현재 클러스터(k)에 속하는 데이터 마스크 설정
    class_member_mask = (db_labels == k)
    # 현재 클러스터(k)에 속하는 데이터 포인트 추출
    xy = X_scaled[class_member_mask]
    plt.plot(xy[:, 0], xy[:, 1], 'o', markerfacecolor=tuple(col), markeredgecolor='k', markersize=6)
 
plt.title('DBSCAN Clustering')
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.show()
```

### 하이퍼파라미터

```python
from sklearn.base import BaseEstimator, ClusterMixin
from sklearn.metrics import silhouette_score

class DBSCANWrapper(BaseEstimator, ClusterMixin):
    def __init__(self, eps=0.5, min_samples=5):
        self.eps = eps
        self.min_samples = min_samples

    def fit(self, X, y=None):
        self.dbscan_ = DBSCAN(eps=self.eps, min_samples=self.min_samples)
        self.dbscan_.fit(X)
        self.labels_ = self.dbscan_.labels_
        return self

    def fit_predict(self, X, y=None):
        self.fit(X)
        return self.labels_
    
def silhouette_scorer(estimator, X):
    labels = estimator.fit_predict(X)
    # 클러스터가 1개 이하인 경우 평가 점수를 -1로 설정
    if len(set(labels)) <= 1:
        return -1
    return silhouette_score(X, labels)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.preprocessing import StandardScaler
import numpy as np

# 데이터 생성 및 스케일링
X, labels_true = make_blobs(n_samples=750, centers=centers, cluster_std=0.4, random_state=0) 
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

#중심점 설정 > 데이터 생성을  위한 > dbscan에 설정하는 것은 아님. 
centers = [[1, 1], [-1, -1], [1, -1]] 

# 하이퍼파라미터 그리드 설정
param_grid = {
    'eps': np.arange(0.1, 1.5, 0.1),
    'min_samples': range(3, 10)
}

# GridSearchCV 설정
grid_search = GridSearchCV(
    estimator=DBSCANWrapper(),
    param_grid=param_grid,
    scoring=silhouette_scorer,
    cv=[(slice(None), slice(None))]  # 비지도 학습에서는 전체 데이터를 사용
)

# 그리드 탐색 수행
grid_search.fit(X_scaled)

# 최적의 하이퍼파라미터 출력
print(f'Best params: {grid_search.best_params_}, Best Silhouette Score: {grid_search.best_score_}')
```

```python
# 최적의 하이퍼파라미터로 DBSCAN 수행
best_eps = grid_search.best_params_['eps']
best_min_samples = grid_search.best_params_['min_samples']
dbscan = DBSCAN(eps=best_eps, min_samples=best_min_samples)
labels = dbscan.fit_predict(X_scaled)

# 클러스터링 결과 시각화
unique_labels = set(labels)
colors = plt.cm.Spectral(np.linspace(0, 1, len(unique_labels)))

for k, col in zip(unique_labels, colors):
    class_member_mask = (labels == k)
    xy = X_scaled[class_member_mask]
    plt.plot(xy[:, 0], xy[:, 1], 'o', markerfacecolor=tuple(col), markeredgecolor='k', markersize=6)

plt.title('DBSCAN Clustering with Optimized Parameters')
plt.show()
```
